```python
import pandas as pd

sec_df = pd.read_csv("data/security_issues.csv")
non_sec_df = pd.read_csv("data/non-security_issues.csv")
```

```python
df = pd.concat([sec_df, non_sec_df]).reset_index(drop=True)
```

```python
df
```

```python
# count unique repositories
df["repository"].nunique()
```

```python
# count number of entries for each repository
df["repository"].value_counts()
```

```python
(df["repository"].value_counts() == 1).sum()
```

```python
# plot number of entries for each repository
df["repository"].value_counts().plot(kind="bar", figsize=(40, 10), title="Number of issues per repository")
```

```python
# count unique repositories
sec_df["repository"].nunique()
```

```python
# plot number of entries for each repository
sec_df["repository"].value_counts().plot(kind="bar", figsize=(40, 10), title="Number of security-related issues per repository")
```

```python
(sec_df["repository"].value_counts() < 10).sum()
```

```python
# count unique repositories
non_sec_df["repository"].nunique()
```

```python
# plot number of entries for each repository
non_sec_df["repository"].value_counts().plot(kind="bar", figsize=(40, 10), title="Number of non-security issues per repository")
```

```python
sec_df["repository"].value_counts().plot(kind="bar", figsize=(40, 10))
non_sec_df["repository"].value_counts().plot(kind="bar", figsize=(40, 10))
```

```python
import matplotlib.pyplot as plt

# set figure size
plt.figure(figsize=(10, 40))

# veritcally align x-axis labels
# plt.xticks(rotation=90)

# plot sec_df["repository"].value_counts() and non_sec_df["repository"].value_counts() in the same bar plot
plt.barh(sec_df["repository"].value_counts().index, sec_df["repository"].value_counts(), label="security-related")
plt.barh(non_sec_df["repository"].value_counts().index, non_sec_df["repository"].value_counts(), label="non-security")

# add legend
plt.legend()
```

```python
# show perentage of security-related issues for each repository
sec_issue_percentage = sec_df["repository"].value_counts() / df["repository"].value_counts()
# sort values
sec_issue_percentage.sort_values(ascending=False, inplace=True)
```

```python
import matplotlib.pyplot as plt

# set figure size
plt.figure(figsize=(10, 40))

# veritcally align x-axis labels
# plt.xticks(rotation=90)

plt.barh(sec_issue_percentage.index, sec_issue_percentage, label="Percentage of security-related issues")

# add legend
plt.legend()
```
